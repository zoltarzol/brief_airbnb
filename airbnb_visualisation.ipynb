```python
df_preparation =df_air_bnb_anvers[["room_type","amenities"] ]
#Je prÃ©pare une dataframe adaptÃ© pour visualiser les relations entre les types de piÃ¨ces et les amÃ¨nagements
df_preparation["amenities"]=df_preparation.amenities.str.count(",")+1
#Ici , j'utilise une astuce comme les amÃ¨nagements sont tous supÃ©rieurs Ã  1 via la vÃ©rification utilisÃ©s plus tÃ´t , je sais qu'il y a des virgules et
#autant de mots que de virgules.Je vais donc comptÃ© le nombre de virgules pour avoir le nombre d'amÃ¨nagements.
df_preparation =df_preparation[["room_type","amenities"] ]
df_groupby=df_preparation.groupby(by="room_type").apply(lambda s: pd.Series({ 
    "amenities_mean": s["amenities"].mean(),
     "amenities_std": s["amenities"].std()

}))
#J'effectue un groupby pour calculer le nombre moyens d'amÃ¨nagement et l'ecart type des amÃ¨nagement en fonction du types de piÃ¨ces.
df_groupby
```

```python
df_preparation.room_type.unique()
```

```python


plt.figure(figsize=(12,6))

amenagement= df_preparation.room_type.unique()
errorValues = [1, 0.5, 2.5, 3]

plt.bar(amenagement, df_groupby.amenities_mean)
plt.errorbar(np.arange(len(amenagement)), df_groupby.amenities_mean, yerr = errorValues,
    fmt = 'none', capsize = 10, ecolor = 'red', elinewidth = 2, capthick = 8)
plt.show()
```

```python
import re
df_annonce = df_air_bnb_anvers.bathrooms_text
df_annonce.unique()
#On crÃ©e des algorithmes qui vont sÃ©parer les informations du 'bathrooms_text'
def first(s):

    return(re.findall("\d+\.\d+|\d+", s))
#first permet de recupere le nombres donnÃ©es
def last(s):
    return(re.split('(\d+)', s)[-1])
#last permet de sÃ©parer le texte 

df_annonce = df_annonce.apply(lambda s: pd.Series({ 
    "nombres_de_baignoires": first(s),
    "caracteristiques" : last(s)

}))
df_annonce_modifie =df_annonce.caracteristiques.str.replace('private bath','2').str.replace('private baths','2').str.replace('Private half',"1").str.replace('shared bath','0.5').str.replace('shared baths','0.5').str.replace('Half-bath','0.5').str.replace("Shared half","0.25").str.replace('bath','1').str.replace('baths','1').str.replace('s','').str.replace('-1',"")
df_annonce_modifie = df_annonce_modifie.astype(float)
df_annonce_finale = pd.DataFrame(df_annonce,columns=["nombres_de_baignoires"])
df_annonce_finale["caracteristique_nombre"]=df_annonce_modifie
df_baignoire = pd.DataFrame([x for x in df_annonce_finale["nombres_de_baignoires"]],columns=["nombre_de_baignoire"])
#On transforme la "liste" en nombre pour prÃ©parer les calculs
df_annonce_finale["nombres_de_baignoires"]=df_baignoire["nombre_de_baignoire"].astype(float)
df_annonce_finale =[x*y for x,y in zip(df_annonce_finale.nombres_de_baignoires , df_annonce_finale.caracteristique_nombre) ] 
#On transforme la dataframe en liste qui multiplie les valeurs via des liste par comprÃ©hension
df_annonce_finale= pd.DataFrame(df_annonce_finale,columns=["valeur_baignoire"])
#On transforme la liste en dataframe pour la prÃ©senter
df_annonce_finale = df_annonce_finale.valeur_baignoire.value_counts()
df_annonce_finale
```

```python


plt.figure(figsize=(15,6))

baignoire = list(df_annonce_finale)
x_pos = np.arange(len(baignoire))

plt.bar(x_pos,baignoire )


plt.show()
```
