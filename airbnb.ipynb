Question n¬∞6
Combien y-a-t-il d‚Äôannonces par valeur de salle de bain?

Analyse et transformation des donn√©es

```python
df_annonce = df_air_bnb_anvers.bathrooms_text
df_annonce.unique()
#On observe la partie du dataframe qui nous int√©resse , soit le texte descriptif des baignoires.
```

```python

#On cr√©e des algorithmes qui vont s√©parer les informations du 'bathrooms_text'
def first(s):

    return(re.findall("\d+\.\d+|\d+", s))
#first permet de recupere le nombres donn√©es
def last(s):
    return(re.split('(\d+)', s)[-1])
#last permet de s√©parer le texte 
```

```python
df_annonce = df_annonce.apply(lambda s: pd.Series({ 
    "nombres_de_baignoires": first(s),
    "caracteristiques" : last(s)

}))
df_annonce
#En appliquant les fonctions pr√©cedentes , on s√©pare en deux colonnes les information via un apply
```

```python
df_annonce_modifie =df_annonce.caracteristiques.str.replace('private bath','2').str.replace('private baths','2').str.replace('Private half',"1").str.replace('shared bath','0.5').str.replace('shared baths','0.5').str.replace('Half-bath','0.5').str.replace("Shared half","0.25").str.replace('bath','1').str.replace('baths','1').str.replace('s','').str.replace('-1',"")
#On applique une transformation , rempla√ßant le texte en nombre que l'on peut manipuler
```

```python
df_annonce_modifie
```

```python
df_annonce_modifie.value_counts()
#Je compte le nombres d'it√©rations des valeurs pour v√©rifier si il y a des valeurs √©tranges
```

```python
df_annonce_modifie = df_annonce_modifie.astype(float)
#On change le type des √©lements pour pr√©parer les calculs
df_annonce_modifie
```

```python
df_annonce_finale = pd.DataFrame(df_annonce,columns=["nombres_de_baignoires"])
df_annonce_finale["caracteristique_nombre"]=df_annonce_modifie
#On organise les donn√©es modifi√©s dans un dataframe pour mieux les pr√©senter et les manipuler plus facilement
df_annonce_finale
```

```python
df_baignoire = pd.DataFrame([x for x in df_annonce_finale["nombres_de_baignoires"]],columns=["nombre_de_baignoire"])
#On transforme la "liste" en nombre pour pr√©parer les calculs
df_baignoire
```

```python
df_baignoire.dtypes
#On observe que les donn√©es sont des strings et non des int/float
```

```python
df_annonce_finale["nombres_de_baignoires"]=df_baignoire["nombre_de_baignoire"].astype(float)
#On transforme les strings en float pour les calculs
```

```python
df_annonce_finale.dtypes
#On v√©rifie
```

```python
df_annonce_finale.nombres_de_baignoires.value_counts()
#On regarde les valeurs pour s'assurer qu'il n y a pas de valeurs √©tranges
```

```python
df_annonce_finale =[x*y for x,y in zip(df_annonce_finale.nombres_de_baignoires , df_annonce_finale.caracteristique_nombre) ] 
#On transforme la dataframe en liste qui multiplie les valeurs via des liste par compr√©hension
df_annonce_finale= pd.DataFrame(df_annonce_finale,columns=["valeur_baignoire"])
#On transforme la liste en dataframe pour la pr√©senter
df_annonce_finale
```
